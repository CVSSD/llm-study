```python
import torch
import math
import numpy as np
import time
```

Vectornization for speed


```python
a = torch.ones(1 ** 10000)
b = torch.ones(1 ** 10000)
c = torch.zeros(1 ** 10000)

start = time.time()
for i in range(1**10000):
    c[i] = a[i] + b[i]
end = time.time()
print('Time cost', end - start)

start = time.time()
c = a + b
end = time.time()
print('Time cost', end - start)
```
