```python
import torch

print('torch version:',torch.__version__)
print('Cuda availability:',torch.cuda.is_available())

if torch.cuda.is_available():
    print('CPU device:',torch.cuda.get_device_name())
    print(torch.cuda.memory_summary())

```

```python
x_float32 = torch.randn(
    10000,
    10000,
    dtype=torch.float32,
    device="cuda:0"
)
print(x_float32.dtype)
print('Memory used:',torch.cuda.memory_allocated()/1024**3,'GB')

x_float16 = x_float32.to(torch.float16)
del x_float32
print(x_float16.dtype)
print('Memory used:',torch.cuda.memory_allocated()/1024**3,'GB')

x_int16 = x_float16.to(torch.int8)
del x_float16
print(x_int16.dtype)
print('Memory used:',torch.cuda.memory_allocated()/1024**3,'GB')
```
