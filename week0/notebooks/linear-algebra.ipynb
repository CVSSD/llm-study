Sum and cumulative sum

```python
x = torch.arange(2 * 3 * 4).reshape(2, 3, 4)
print(x)
print(x.sum(axis = 0))
print(x.sum(axis = 1))
print(x.sum(axis = 2))

sum_x = x.sum(axis = 0, keepdim=True)
print(sum_x)
print(sum_x.shape)
cumsum_x = x.cumsum(axis = 2)
print(cumsum_x)
print(cumsum_x.shape)
```

```python
import time
A = torch.randn( (2 ** 10, 2 ** 16) )
B = torch.randn( (2 ** 16, 2 ** 5) )
C = torch.randn( (2 ** 5, 2 ** 16) )

start = time.time()
A @ B
end = time.time()
print("time spent", end - start)
start = time.time()

start = time.time()
A @ C.T
end = time.time()
print("time spent", end - start)

C = B.T
print(C.data_ptr())
print(B.data_ptr())
print(C.data_ptr() == B.data_ptr())
start = time.time()
A @ C.T
end = time.time()
print("time spent", end - start)
start = time.time()
```

Both (AB) and (AC^T) require the same number of floating-point operations because the matrix dimensions are identical. If (B) and (C) are independent matrices initialized separately, high-performance libraries such as BLAS or cuBLAS can efficiently handle transposed operands, so their execution times are usually very similar.

If (C = B^T) is created **without cloning memory**, then (C) is only a transposed **view** of (B). It shares the same underlying storage but has different strides. Consequently,

[
C^T = B,
]

so computing (A C^T) is effectively the same as computing (AB), without creating an additional copy of the matrix. This reduces memory usage and avoids the cost of copying data. The transpose operation itself is essentially free because it only changes the tensor's metadata (shape and strides), not the underlying data.
