```python
x = torch.randn(5, 4, dtype=torch.float32)
print(x)
x_ptr = x.data_ptr()
print("Array start adress",x_ptr)
```

The addresss of element in row 0

```python
for i in range(x.size(1)):
    print(x[0,i].data_ptr())
```

The address of element in col 0

```python
for i in range(x.size(0)):
    print(x[i,0].data_ptr())
```

```python
x = x.reshape(1, -1)
print(x)
for i in range(x.numel()):
    print(x[0, i].data_ptr())
```

```python
a = torch.arange(4).reshape(2,2)
b = torch.arange(2).reshape(1,2)
c = torch.arange(2).reshape(2,1)

print(a + b)
print(a + c)
```

To save memory, we would want to the tense still using the same memory after operation.

```python
address = id(a)
print("address before operstion:",address)
a = a + b
print("address after operstion:",id(a))

address = id(a)
print("address before operstion:",address)
a += b
print("address after operstion:",id(a))

address = id(a)
print("address before operstion:",address)
a[:] = a + b
print("address after operstion:",id(a))

```
