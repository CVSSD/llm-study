```python
import torch 
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import numpy as np

def f(x):
    return 3 * x ** 2 - 4 * x

def numerical_lim(f, x, h):
    return (f(x + h) - f(x)) / h

h = 0.1
for i in range(5):
    print(f'h={h:.5f}, numerical limit={numerical_lim(f, 1, h):.5f}')
    h *= 0.1
```

```python
def plot(x, y, title=None, xlabel='x', ylabel='y',
         xsclar='linear', ysclar='linear', figsize=(6,4),
         grid=True):
    plt.figure(figsize=figsize)
    plt.plot(x, y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)

    plt.gca().set_xscale(xsclar)
    plt.gca().set_yscale(ysclar)

    if title:
        plt.title(title)
    
    if grid:
        plt.grid(True)
        plt.gca().xaxis.set_major_locator(MultipleLocator(2))
        plt.gca().yaxis.set_major_locator(MultipleLocator(20))

    plt.show()
    
```

```python
x = torch.arange(10)
plot(x, f(x))
```
